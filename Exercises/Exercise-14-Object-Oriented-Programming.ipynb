The objective of these classes is to be able to calculate the area and perimeter of any instantiation of the above quadrilaterals. 

Your design should utilise inheritance and polymorphism to ensure maximal code-reuse.

Now implement all the above classes.


````python
#YOUR_CODE_HERE
import math 

class trapezium():
    def __init__(self,base1, base2, side1, side2): #base1, base2 are parallel sides (base1 > base2)
        self.base1 = base1
        self.base2 = base2
        self.side1 = side1
        self.side2 = side2

    def perimeter(self):
        return self.base1 + self.base2 + self.side1 + self.side2
    
    def area(self):
        #Cut the trapezium into a parallelogram and triangle
        #s is half of the diameter of the triangle
        s = (self.side1 + self.side2 + self.base1 - self.base2)/2
        #Obtain area of triangle using Heron's Formula
        triangle = math.sqrt(s*(s - self.side1)*(s - self.side2)*(s-self.base1+self.base2)) 
        #h is the height of the trapezium
        h = (triangle*2) / (self.base1-self.base2)
        #Find the area of trapezium using formula
        return ((self.base1+self.base2)/2) * h

class parallelogram(trapezium):
    #height < sides
    def __init__(self,base1,base2,side1,side2, height): #Identical bases and sides (Height needed for area)
        self.base1 = base1
        self.base2 = base2
        self.side1 = side1
        self.side2 = side2
        self.height = height

    def area(self):
        return self.base1 * self.height
    
class kite(trapezium):
    #Diagonals needed for area
    def __init__(self,base1,base2,side1,side2, diagonal1,diagonal2): #identical bases and sides 
        self.base1 = base1
        self.base2 = base2
        self.side1 = side1
        self.side2 = side2
        self.diagonal1 = diagonal1
        self.diagonal2 = diagonal2

    def area(self):
        return (self.diagonal1*self.diagonal2) / 2
    
class rhombus(kite):
    pass

class rectangle(trapezium):
    #Assume bases and sides are identical
    def area(self):
        return self.base1 * self.side1
    
class square(rectangle):
    pass



#Trapezium test
shape1 = trapezium(25,10,13,14)
print(shape1.perimeter())
print(shape1.area())    #Area is 196cm
        
#Parallelogram test
shape2 = parallelogram(10,10,5,5,4)
print(shape2.perimeter())
print(shape2.area())

#Square test
shape3 = square(5,5,5,5)
print(shape3.perimeter())
print(shape3.area())
````

In mathematics, a complex number $z$ is a number that can be expressed in the form $$x+iy$$ where $x,y\in \mathbb{R}$ and $i$ is a symbol called the **imaginary unit** such that $i^2=-1$. For example, $2+3i$, $-5-i$, $i=0+1i$, $1.5=1.5+0i$ are all complex numbers.

If $z=x+iy$ is a complex number, $x$ is called the real part of $z$ and is denoted as $\text{Re}(z)$, while $y$ is called the imaginary part of $z$ and is denoted as $\text{Im}(z)$. 

A complex number $z=a+ib$ can thus be identified with an ordered pair $(a,b)$ of real numbers, which in turn may be interpreted as coordinates of a point in a two-dimensional space. Thus, we can plot complex numbers on a Cartesian plane where the horizontal (labelled as $\text{Re}$) axis is generally used to display the real part, with increasing values to the right, and the imaginary part marks the vertical (labelled as $\text{Im}$) axis, with increasing values upwards.

Since a complex number is a point on a 2D plane, it makes sense to talk about the distance of the complex number $z$ from the origin. 

If $z=x+iy$, then the distance of the complex number number from the origin is called the *modulus of $z$*, denoted as $|z|$ and thus, $$|z|=\sqrt{x^2+y^2}.$$

Furthermore, the angle subtended between the positive real axis and the line segment $Oz$ in a **counterclockwise sense**, is called the *argument of $z$* and is denoted as $\text{arg}(z)$. Note that we will define $-\pi<\arg(z)\leq \pi$. In other words, if the complex number is located in quadrant 3 and 4, the argument of the complex number will be a negative value.

Implement the class `Complex` with the following specification.

>```
>|  Complex                                |

>|  constructor(real: FLOAT, imag: FLOAT)  |

This class stores a complex number.

The class methods work as follows:

>```
>| Method                            | Description                                               |
>|-----------------------------------|-----------------------------------------------------------|
>| constructor(re: FLOAT, im: FLOAT) | The initialisation method                                 |
>| Re(): COMPLEX                     | Returns the real part of the complex number               |
>| Im(): COMPLEX                     | Returns the imaginary part of the complex number          |
>| mod(): FLOAT                      | Returns the modulus of the complex number                 |
>| arg(): FLOAT                      | Returns the argument of the complex number                |
>| __str__(): STRING                 | Returns the string representation of the complex number   |
>|                                   | in the form `<real_part>+i*<imaginary_part>`              |
>|-----------------------------------|-----------------------------------------------------------|
>```

````python
#YOUR_CODE_HERE
import math

class Complex():
    #Task 1
    def __init__(self,real: float , imag: float):
        self.real = real
        self.imag = imag
        
    def Re(self):
        return self.real
    
    def Im(self):
        return f"{self.imag}i"
    
    def mod(self):
        return math.sqrt(self.real**2 + self.imag**2)
    
    def arg(self):
        #Guard cases for certain scenarios
        if self.real == 0:
            if self.imag < 0:
                return -(math.pi/2)
            else:
                return math.pi/2
        if self.imag == 0:
            if self.real < 0:
                return math.pi
        #Quadrants 2 or 3
        if self.real < 0:
            #Quadrant 3
            if self.imag < 0:
                return math.atan(self.imag / self.real) - math.pi
            else:
                return math.atan(self.imag / self.real) + math.pi
        #Quadrant 1 or 4
        return math.atan(self.imag / self.real)

        return math.atan(self.imag / self.real)   #Returns in radians

    def __str__(self):
        return f"{self.real} + {self.imag}i"
    
    #Task 2
    def __add__(self, c : complex):
        return f"{self.real + c.real} + {self.imag + c.imag}i"
    
    def __sub__(self, c: complex):
        return f"{self.real - c.real} + {self.imag - c.imag}i"
    
    def __mul__(self, c: complex):
        return f"{self.real*c.real - self.imag*c.imag} + {(self.real*c.imag + c.real*self.imag)}i"
    
    def __truediv__(self, c:complex):
        const = 1/(c.real**2 + c.imag**2)
        return f"{const*(self.real*c.real + self.imag*c.imag)} + {const*(c.real*self.imag - self.real*c.imag)}i" 
        
    
number = Complex(1,2)
print(number)
print(number.Re())
print(number.Im())
print(number.mod())
print(number.arg())

#Task 2
v = Complex(3,4)
print(number.__add__(v))
print(number.__sub__(v))
print(number.__mul__(v))
print(number.__truediv__(v))

````

````python
#YOUR_CODE_HERE
import math

class ThreeDVector():
    def __init__(self, a, b, c):
        self.x = a
        self.y = b
        self.z = c

    #To obtain the tuple value (Needed for calculations)
    def getvec(self):
        return (self.x, self.y, self.z)
    
    def __str__(self):
        return f"{(self.x, self.y, self.z)}"
    
    #Note d represents another vector
    def __vec_add__(self, d):
        return (self.x + d[0], self.y + d[1], self.z + d[2])
    
    def __dot__(self, d):
        return (self.x * d[0]) + (self.y * d[1]) + (self.z * d[2])
    
    def __cross__(self,d):
        return (self.y*d[2] - self.z*d[1], self.z*d[0] - self.x*d[2], self.x*d[1] - self.y*d[0])
    
    def __neg__(self):
        return (-self.x, -self.y, -self.z)
    
    def __mod__(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)
    
    def __unit__(self):
        return (self.x / self.__mod__(), self.y / self.__mod__(), self.z / self.__mod__())

    
vec1 = ThreeDVector(1,2,3)
vec2 = ThreeDVector(4,5,6)
print(vec1)
print(vec2)
print(vec1.__vec_add__(vec2.getvec()))
print(vec1.__dot__(vec2.getvec()))
print(vec1.__cross__(vec2.getvec()))
print(vec1.__neg__())
print(vec1.__mod__())
print(vec1.__unit__())


````

Implement the class `LONGINT` with the following specification.

>```
>| LONGINT                        |
>|--------------------------------|
>|  value: STRING                 |
>|  sign: INTEGER                 |
>|--------------------------------|
>|  constructor(STRING)           |
>|  getValue(): STRING            |
>|  setValue(s:STRING)            |
>|  add(l:LONGINT): LONGINT       |
>|  subtract(l:LONGINT): LONGINT  |
>|  multiply(l:LONGINT): LONGINT  |
>|  divide(l:LONGINT): FLOAT      |
>|  quotient(l:LONGINT): LONGINT  |

>|  power(l:LONGINT): LONGINT     |
>|  factorial(): LONGINT          |
>|--------------------------------|

This class stores a signed integer of any length.

The class methods work as follows:

>```
>| Method                        | Description                                                       |
>|-------------------------------|-------------------------------------------------------------------|
>| constructor()                 | The initialisation method                                         | 
>| getValue(): STRING            | Reassigns the signed value of **this**                            |
>| add(l:LONGINT): LONGINT       | Returns the addition of the given `LONGINT` and `self`            |
>| subtract(l:LONGINT): LONGINT  | Returns the subtraction of `LONGINT` from `self`                  |
>| multiply(l:LONGINT): LONGINT  | Returns the product of the given `LONGINT` and `self`             |
>| divide(l:LONGINT): FLOAT      | Returns `self` divided by the given `LONGINT`                     |
>| quotient(l:LONGINT): LONGINT  | Returns the quotient from `self` divided by the given `LONGINT`   |
>| modulo(l:LONGINT): LONGINT    | Returns the remainder from `self` divided by the given `LONGINT`  |
>| power(l:LONGINT): LONGINT     | Returns this to the power of the given `LONGINT`                  |
>| factorial(): LONGINT          | Returns factorial of `self`                                       |
>|-------------------------------|-------------------------------------------------------------------|
>```

````python
#YOUR_CODE_HERE
#Assumption: sign represents +,- , therefore type is a string
#Value represents number, therefore type is an integer

class Longint():
    def __init__(self, sign: str, value : int):
        self.value = value
        self.sign = sign

    def getValue(self):
        if self.sign == "-":
            return -self.value
        else:
            return self.value
        
    def setValue(self, newvalue):
        self.value = newvalue

    def __add__(self, l: int):
        return self.getValue() + l
    
    def __sub__(self, l: int):
        return self.getValue() - l
    
    def __mul__(self, l: int):
        return self.getValue() * l
    
    def __div__(self, l: int):
        return self.getValue() / l
    
    def __quo__(self, l: int):
        return self.getValue() // l
    
    def __rem__(self, l: int):
        return self.getValue() % l
    
    def __pow__(self, l: int):
        return self.getValue() ** l
    
    def __factorial__(self):
        if self.sign == "+":
            total = 1
            for x in range(1 , self.value + 1):
                total *= x
        else:
            return -1
        return total

number = Longint("+", 5)
number2 = Longint("-", 3)
print(number.getValue())
print(number.__add__(number2.getValue()))
print(number.__factorial__())
````

````python
#YOUR_CODE_HERE
import csv
import random

class maze():
    def __init__(self):
        self.grid = []
    
    def reset(self):
        with open("MAZE.TXT") as m:
            data = csv.reader(m)
            for row in data:
                self.grid.append(row[0])

    def clear(self):  #Clear default P and O placements
        for n in range(len(self.grid)):
            for i in range(len(self.grid[n])):
                if self.grid[n][i] == 'P' or self.grid[n][i] == 'O':
                    self.grid[n] = self.grid[n][:i] + "." + self.grid[n][i+1:]
    
    def clearp(self): #Only get rid of player
        for n in range(len(self.grid)):
            for i in range(len(self.grid[n])):
                if self.grid[n][i] == 'O':
                    self.grid[n] = self.grid[n][:i] + "." + self.grid[n][i+1:]


    def posindex(self, x , y): #This function asks for coordinates for a position and returns whats on it
        return self.grid[y][x]

    def display(self):
        for i in self.grid:
            temprow = ""
            for x in i:
                temprow += x
                temprow += " "
            print(temprow)

    #Task 2
    def addprize(self):
        #Randomize spot for a potential prize placement
        while True:
            px = random.randint(0,9)
            py = random.randint(0,10)
            if self.grid[py][px] == "X":
                pass
            else:
                self.grid[py] = self.grid[py][:px] + "P" + self.grid[py][px+1:]
                break

    #Task 3
    def addplayer(self, playero):
        ox = playero.get_xpos()
        oy = playero.get_ypos()
        if self.grid[oy][ox] == "X":
            print("Cannot be placed here")
        else:
            self.grid[oy] = self.grid[oy][:ox] + "O" + self.grid[oy][ox+1:]



        
    
maze1 = maze()
maze1.reset()
````
